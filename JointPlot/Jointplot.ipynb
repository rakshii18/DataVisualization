## Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Importing the dataset

```python
df=pd.read_csv('tempYearly.csv')
df
```

### Heatmap

```python
sns.set(rc={'figure.figsize':(20,25)})
sns.jointplot("Temperature", "Rainfall", data=df)
plt.show()
```

```python
sns.set(rc={'figure.figsize':(20,25)})
sns.jointplot("Temperature", "Rainfall", data=df, kind="hex")
plt.show()
```

```python
#Regression Plots
sns.set(rc={'figure.figsize':(20,25)})
sns.jointplot("Temperature", "Rainfall", data=df,kind='reg')
plt.show()
```
