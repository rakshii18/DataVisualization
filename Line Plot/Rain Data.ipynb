# Importing the required libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
os.chdir('.\Desktop\Data')
```

### Importing the DataSet - Rain DataSet

```python
df=pd.read_json('./rain.json')
df
```

### Plotting the data

```python
#Month v/s Temperature
plt.figure(figsize=(15,5)) #width, height
df.plot(x='Month',y='Temperature', label='Temperature')
plt.show()
```

```python
#Month v/s rainfall
plt.figure(figsize=(17,5))
df.plot(x='Month', y='Rainfall', label="Rainfall")
plt.show()
```

```python
#Combining both plots into 1.
plt.figure(figsize=(17,5))
plt.plot(df["Month"],df["Temperature"], label="Temperature")
plt.plot(df["Month"],df["Rainfall"], label="Rainfall")
plt.legend()
plt.show()
```
